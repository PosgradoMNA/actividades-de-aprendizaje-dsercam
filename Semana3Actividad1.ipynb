#### Fundamentos de base de datos

**1.	Fundamentos de bases de datos y para ciencia de datos.**  <br>
<p align="justify">
Una base de datos es un conjunto de datos lógicamente relaciones, cuyo objetivo es soportar las actividades y decisiones de una organización.
Una característica importante de una base de datos es que éstas son autodescriptivas, donde sus metadatos junto con su esquema nos ayudan a entender las relaciones y los atributos de sus múltiples entidades.<br><br>
Respecto a la ciencia de datos, las bases de datos son una de una de diferentes fuentes de información, y aunque normalmente contienen información estructurada, algunas de sus entidades y atributos también puedes servir de repositorio para datos semi estructurados y no estructurados.<br><br>
Para un analista, las bases de datos también son un punto de entrada para conocer un dominio, pues al observar sus distintas entidades, atributos y la información contenida en ellas, esto nos ayuda a realizar conjeturas y formular preguntas adecuadas para entender la información adecuadamente, posteriormente procesarla de forma adecuada para su uso.<br><br>
Entender conceptos como: entidades, tablas, relaciones, normalización, atributos o campos, registros, etc. es en extremo relevante, pues dichos conceptos se mantienen más allá de la tecnología-herramienta DBMS usada. </p>
<br><br>

**2.	Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.** <br>
<p align="justify">
Los almacenes de datos conjuntan información de distintas fuentes, pudiendo ser éstas sistemas de bases de datos transaccionales -OLPTP- carga de información a través de archivos con información estructurada, semi estructurados, o no estructurados, APIs, o incluso otros sistemas de datos orientados a procesos analíticos – OLTP-. El objetivo principal de un almacén de datos es soportar los procesos de análisis de datos y BI para la toma de decisiones.<br><br>
Una de las características más relevantes de un almacén de datos es que estos utilizan un nivel de normalización distinto al de bases de datos transaccionales -las cuales suelen estar diseñadas para reducir la redundancia y optimizar transacciones- y están en lugar optimizadas para procesos de análisis. Los almacenes de datos usan modelos multi-dimensionales para almacenar la información, por ejemplo, en modelos estrella (star schemas) dónde las tablas de hechos están rodeadas por distintas tablas de dimensiones para así poder describir los eventos guardados en las tablas de hechos.
Los DWs también suelen almacenar datos en sus múltiples formatos, estructurados, semi estructurados y no estructurados.<br><br>
Al trabajar con un DW, uno de los aspectos más importantes a conocer es su método de carga ya sea ETL o ELT. Los procesos de ETL son el antecedente histórico de los ELTs, cuando el proceso de transformación se realiza antes de la carga al DW, actualmente, los ELTs hacen uso de los clusters de información y computo para realizar la transformación después de la carga y así aprovechar el poder computacional de las nuevas tecnologías para la fase de transformación de datos, pues esta es la más demandante y compleja de escalar.</p>

Definción de los atributos:
Attribute Information:

**Y:** Variable de respuesta donde: default payment (Yes = 1, No = 0) <br> 
**X1:** Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit. <br> 
**X2:** Gender (1 = male; 2 = female). <br> 
**X3:** Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). <br> 
**X4:** Marital status (1 = married; 2 = single; 3 = others). <br> 
**X5:** Age (year). <br> 
**X6 - X11:** History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above. <br> 
**X12-X17:** Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005. <br> 
**X18-X23:** Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005.
