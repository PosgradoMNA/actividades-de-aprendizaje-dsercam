### PCA
Realizaremos dos PCA, en uno usaremos un data frame donde los datos han sido normalizados y en otro un dataframe donde no se ha aplicado la normalización, esto nos ayudará a identificar la importancia de la normalización para un uso adecuado del PCA.

- Utilizaremos objeto _pcs_ para aplicar PCA al dataframe sin normalización
- Utilizaremos objeto _pcsScaled_ para aplicar PCA en el data frame normalizado.

```python
#importamos librerias necesarias.
import sklearn
from sklearn.decomposition import PCA #uso del PCA
from sklearn import preprocessing # para normalizar el dataframe

# fit PCA en frame de datos sin normalizar
pcs = PCA()
pcs.fit(processedDataFrame)

#fit PCA en frame de datos al que se le aplica normalización
pcsScaled= PCA()
pcsScaled.fit(preprocessing.scale(processedDataFrame)) # preprocessing.scale: Standardize features by removing the mean and scaling to unit variance.

#Resumen de metricas para PCA sin normalización
summary ={'Standard deviation': np.sqrt(pcs.explained_variance_),
          'Proportion of variance': pcs.explained_variance_ratio_,
          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)}
#Resumen de metricas para PCA con normalización
summaryScaled ={'SCALED - Standard deviation': np.sqrt(pcsScaled.explained_variance_),
          'SCALED - Proportion of variance': pcsScaled.explained_variance_ratio_,
          'SCALED -Cumulative proportion': np.cumsum(pcsScaled.explained_variance_ratio_)}
```

Al ver el resumen de las métricas, podemos ver como el no aplicar la normalización resulta en que algunas variables reflejen más "peso" sobre otras en relación a sus variancias, esto debido a la magnitud de los rangos en cada una de las variables.

La normalización de los datos resulta en varianzas mas equilibradas entre las distintas variables.

```python
pcsSummary = pd.DataFrame(summary).transpose().round(3)
pcsSummary.columns = ['PC{}'.format(i) for i in range(1,len(pcsSummary.columns)+1)]
pcsSummary
```

```python
pcsScaledSummary = pd.DataFrame(summaryScaled).transpose().round(3)
pcsScaledSummary.columns = ['PC{}'.format(i) for i in range(1,len(pcsSummary.columns)+1)]
pcsScaledSummary
```

  Creo que en este caso las dos opciones suenan interasantes, y aconsejaria intentar entrenar el modelo bajo ambas configuraciones, pues ambas opciones reducen un numero importante de dimensiones, y comparar la exactitud de los modelos podría orientarnos hacía una mejor decisión.  <br><br>

- **¿Cuándo se recomienda realizar un PCA y qué beneficios ofrece para Machine Learning?**
Se recomienda usar el PCA cuando contamos con un alto número de dimensiones (features) donde ademas existe depedencia fuerte en entre variables (multicolinealidad). El aplicar el PCA nos ayudará a generar nuevos componentes con baja dependencia entre ellos y ademas tambiar a eliminar el ruido en los datos, pues la idea es usar solo los componentes que describan la mayoría de los datos (los componentes acumulados que otorgen cierto nivel de varianza "describen" los datos origniales.<br><br>
El machine learning se beneficia de este proceos pues las operaciones computacionales para los modelos se reducen en complejidad, el performance aumenta -procesamiento más rápido-. De igual forma los componentes nos ayudan a reducir el overfitting, pues al reducir el ruido y el númeor de variables, el entrenamiento se mantendrá solamente influenciado por las variables mas relevantes.
